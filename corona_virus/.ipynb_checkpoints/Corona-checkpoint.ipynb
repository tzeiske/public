```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter

%matplotlib inline
```

```python
df = pd.read_csv('corona_03-11.csv')
```

```python
df.head()
```

```python
df['Deaths']=df['Deaths']+1
```

```python
df['Deaths']
```

```python
df['health_ratio']=df['Positive']/df['Deaths']
```

```python
df['health_ratio']
```

```python
df.index=df['Country']
```

```python
df.plot.bar(subplots=True)
```

```python
df['health_ratio'].plot.bar(rot=30, width=0.5)
```

```python
x_pos=np.arange(0,10,1)
df = df.sort_values('health_ratio', ascending=False).reset_index(drop=True)
```

```python
ax=df['health_ratio'].plot.bar(width=0.7, rot=40, figsize=(25,15))
#ax.set_xticklabels(df['Country'], rotation=65, ha='center') #, rotation_mode="anchor")


# configure tick labels
ax.set_xticklabels(df['Country'], rotation=45, ha='right', fontsize=18)
ax.set_yticklabels(ax.get_yticks(), fontsize=14)

# configure y axis label
ax.set_ylabel('Health Ratio (cases / deaths)', fontsize=20)
ax.yaxis.set_major_formatter(FormatStrFormatter('%d'))

fig=ax.get_figure()

# set figure title
fig.suptitle('Worldwide COVID-19 Survival', fontsize=22)

# save and display figure
plt.tight_layout(rect=[0, 0, 1, 0.95])

# set figure title
fig.suptitle('Worldwide COVID-19 Survival', fontsize=22)

# save and display figure
plt.tight_layout(rect=[0, 0, 1, 0.95])
fig.savefig('health_ratio.png', dpi=75)
```
